```python
import numpy as np
from matplotlib import pyplot as plt

x=np.arange(1,11)
x

y=2*x
y

plt.plot(x,y)
plt.show()


#adding title and labels
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()


#changing line style
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

#aading two lines in one
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='--',linewidth=2)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()



#adding sub plots
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(1,2,1) #subplot(row,col,index of graph)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)


plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='--',linewidth=2)

plt.show()


#Bar plot
student={"Bob":57,"Denne":98,"Anne":87}
names=list(student.keys())
marks=list(student.values())
plt.bar(names,marks)
plt.show()

#horizontal bar
student={"Bob":57,"Denne":98,"Anne":87}
names=list(student.keys())
marks=list(student.values())
plt.barh(names,marks)
plt.show()


#Box plot
first=[1,2,3,4,5,6,7,8,9]
second=[1,2,3,4,5,4,3,2,1]
third=[4,3,2,7,6,8,1,5,6]

data=list([first,second,third])
plt.boxplot(data)
plt.show()


#Violin plot
first=[1,2,3,4,5,6,7,8,9]
second=[1,2,3,4,5,4,3,2,1]
third=[4,3,2,7,6,8,1,5,6]

data=list([first,second,third])
plt.violinplot(data)
plt.show()


#Pie-Chart
cities=["Mumbai","Delhi","Bangalore","Hyderabad","Chennai"]
quantity=[10,50,30,20,60]

plt.pie(quantity,labels=cities)
plt.show()


#changing colors and adding percentage
cities=["Mumbai","Delhi","Bangalore","Hyderabad","Chennai"]
quantity=[10,50,30,20,60]

plt.pie(quantity,labels=cities,autopct='%0.1f%%',colors=['red','green','yellow','pink','grey'])
plt.show()

#DoughNut chart
cities=["Mumbai","Delhi","Bangalore","Hyderabad","Chennai"]
quantity=[10,50,30,20,60]

plt.pie(quantity,labels=cities,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()







```
